3. Past Usage:
    1. Smith,~J.~W., Everhart,~J.~E., Dickson,~W.~C., Knowler,~W.~C., \&
       Johannes,~R.~S. (1988). Using the ADAP learning algorithm to forecast
       the onset of diabetes mellitus.  In {\it Proceedings of the Symposium
       on Computer Applications and Medical Care} (pp. 261--265).  IEEE
       Computer Society Press.

       The diagnostic, binary-valued variable investigated is whether the
       patient shows signs of diabetes according to World Health Organization
       criteria (i.e., if the 2 hour post-load plasma glucose was at least 
       200 mg/dl at any survey  examination or if found during routine medical
       care).   The population lives near Phoenix, Arizona, USA.

       Results: Their ADAP algorithm makes a real-valued prediction between
       0 and 1.  This was transformed into a binary decision using a cutoff of 
       0.448.  Using 576 training instances, the sensitivity and specificity
       of their algorithm was 76% on the remaining 192 instances.

4. Relevant Information:
      Several constraints were placed on the selection of these instances from
      a larger database.  In particular, all patients here are females at
      least 21 years old of Pima Indian heritage.  ADAP is an adaptive learning
      routine that generates and executes digital analogs of perceptron-like
      devices.  It is a unique algorithm; see the paper for details.

5. Number of Instances: 768

6. Number of Attributes: 8 plus class 

    7. For Each Attribute: (all numeric-valued)
       1. Number of times pregnant
       2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
       3. Diastolic blood pressure (mm Hg)
       4. Triceps skin fold thickness (mm)
       5. 2-Hour serum insulin (mu U/ml)
       6. Body mass index (weight in kg/(height in m)^2)
       7. Diabetes pedigree function
       8. Age (years)
       9. Class variable (0 or 1)

8. Missing Attribute Values: Yes

# Some required packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

# Load dataset

```python
data = pd.read_csv('./dataset/pima-indians-diabetes.csv')
data.head(3)
```

### Add names to columns

```python
columns = [
    'number_pregnant', 
    'glucose_concentration', 
    'blood_pressure', 
    'triceps', 
    'insulin', 
    'bmi', 
    'pedigree', 
    'age', 
    'class'
]
```

```python
data.columns = columns
```

```python
data.head()
```

# Split features and labels

```python
labels = data['class'].as_matrix().astype(np.int32)
feat_data = data.drop('class', axis=1).as_matrix().astype(np.float32)
```

## One-hot encoding

```python
# labels = np.zeros((label_data.shape[0], len(set(label_data))))
```

```python
# for i in range(labels.shape[0]):
#     labels[i, label_data[i]] = 1
```

# Split train test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    feat_data,
    labels,
    test_size=0.3,
    random_state=77
)
```

# Scale the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_x_train = scaler.fit_transform(X_train)
```

```python
scaled_x_test = scaler.transform(X_test)
```

# Keras API

https://www.tensorflow.org/api_docs/python/tf/contrib/keras

```python
import tensorflow as tf
from tensorflow.contrib.keras import models
```

```python
model = models.Sequential()
```

# Add layers to the model

```python
from tensorflow.contrib.keras import layers
```

```python
model.add(layers.Dense(input_dim=8, units=13, activation='relu'))
model.add(layers.Dense(units=10, activation='relu'))
model.add(layers.Dense(units=13, activation='relu'))
```

```python
model.add(layers.Dense(units=2, activation='softmax'))
```

# Compile the model

```python
from tensorflow.contrib.keras import losses, optimizers, metrics
```

```python
model.compile(
    optimizer=optimizers.Adam(lr=1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

# Train the model

```python
model.fit(scaled_x_train, y_train, epochs=500, batch_size=10)
```

# Predictions

```python
predictions = model.predict_classes(scaled_x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test, predictions))
```
